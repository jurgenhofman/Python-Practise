```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#eerst data in dataframe plaatsen en beschouwen
df = pd.read_excel('G:\Data Science & PowerBI\Energie.xlsx')
df.head(5)
```

```python
#zorg dat perdiode een jaartal is
df["Perioden"] = df["Perioden"].str[:4]
df.head(5)
```
